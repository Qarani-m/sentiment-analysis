```python
from google.colab import drive
drive.mount('/content/drive')
file_path  ="/content/drive/MyDrive/dataset.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df['Sentiment_Score'] = df["Tweet Text"].apply(lambda text: TextBlob(text).sentiment.polarity)
df['Sentiment'] = df["Sentiment_Score"].apply(lambda score: "positive" if score > 3 else "negative")
df['Sentiment'] = df["Sentiment"].map({"positive":1, "negative":0})
df.head()
```
